```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 200, centers = 4, cluster_std = 0.5, random_state=0)
print('X前五筆資料:\n',X[:5])
print('y前五筆資料:\n',y[:5])
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(4,3))
plt.scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
# TODO(2)
# 繪製肘部法則圖
from sklearn.cluster import KMeans

kmeans_k = [KMeans(n_clusters=k, init='k-means++', n_init=15, max_iter=200, random_state=0).fit(X) for k in range(1,10)]  #1~9
inertia = [model.inertia_ for model in kmeans_k]

plt.figure(figsize=(8,6))
plt.scatter(range(1, 10), inertia, color='b', marker='o')  # 修正這一行
plt.plot(range(1, 10), inertia, color='b', linestyle='-', linewidth=1)  # 加上線條繪製

# 在每個點旁邊標註數值
for i, val in enumerate(inertia):
    plt.text(i + 1, val, f'{val:.4f}', color='green', fontsize=8, ha='left', va='bottom')

plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=15, max_iter=200, random_state=0).fit(X)

# 查看收斂情況
# inertia_：最後一次迭代的總內部誤差平方和（即群集內的總距離）
# n_iter_：K-Means 在收斂前所執行的實際迭代次數
print("Inertia:\n", kmeans.inertia_)
print("Iterations:\n", kmeans.n_iter_)
# cluster center
cluster_center = kmeans.cluster_centers_
print('Cluster centers:\n', cluster_center)
```
