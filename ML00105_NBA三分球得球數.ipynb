#線性回歸

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/TQC+/data/NBApoints.csv')
df.info()
```

```python
print('Data shape:',df.shape)
df.head(10)
```

```python
# X = Pos(Object) Age(int64) Tm(object)
# y = 3P 三分球得球數

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Pos'] = le.fit_transform(df['Pos'])
df['Tm'] = le.fit_transform(df['Tm'])
print('Pos:',df['Pos'])
print('Tm:',df['Tm'])
```

```python
X = df.loc[:,['Pos','Age','Tm']]
y = df.loc[:,'3P']

print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# ([5, 28, 10]) 的三分球數
data = pd.DataFrame([5, 28, 10]).T
data_pred = model.predict(data)
print(f'三分球數:{data_pred[0]:.4f}')
```

```python
from sklearn.feature_selection import f_regression
# p-values是否小於0.05

r_squared = model.score(X, y)
p_values = f_regression(X, y)[1]
f_stat = f_regression(X, y)[0]

print(f'R-squared:{r_squared:.4f}')
print(f'P-values:{p_values:}')
print('P-values:', ', '.join(f'{p:.4f}' for p in p_values))
print(f'F-statistics:{f_stat}')
```
