```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive/')
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/TQC+/data/pokemon.csv')

print(df.shape)
df.head()
```

```python
# 處理遺漏值:移除'Attack', 'Defense'有遺漏值之寶可夢 (dropna():刪除包含 NaN 的資料列)
features = ['Attack','Defense']
df.dropna(subset=features, inplace = True)
# df = df.dropna(subset=features, inplace = False)
df.head()
```

```python
# 針對Type1裡的'Normal','Fighting','Ghost'寶可夢的'Attack'和'Defense'進行分類
mask = df['Type1'].isin(['Normal','Fighting','Ghost'])
df = df[mask]
print(df.shape)
df.head(10)
```

```python
X = df.loc[:,features]
y = df.loc[:,'Type1']

print(X)
print(y)
```

```python
# 編碼 Type1:LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

print(y)
```

標準化:
X'=(X−μ)/σ
X'=標準化後的數值
X=原始數值
μ=平均值
σ=標準差


```python
# 特徵標準化  StandardScaler:均值=0,標準差=1
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)
```

# **線性支援向量分類器(Support Vector Classifier,SVC)**

*   C=0.1：這是正則化參數。C 值越小，模型越保守，正則化效果越強；C 值越大，模型對訓練資料的擬合越好，但可能會過擬合。設定 C=0.1 意味著我們希望有較強的正則化，以防止過擬合。

*   dual=False：這個參數決定是否要用對偶 (dual) 形式來解問題。當樣本數量大於特徵數量時，設置 dual=False 會更有效率。如果樣本數量小於特徵數量，使用對偶形式 (dual=True) 會更合適。
*   class_weight='balanced'：這個參數用來處理類別不平衡問題。當設置為 balanced 時，LinearSVC 會根據訓練集中每個類別的頻率來自動調整各個類別的權重，以平衡類別的影響。








```python
# 建立線性支援向量分類器，C=0.1, dual=False, class_weight='balanced'
from sklearn.svm import LinearSVC
svc_clf = LinearSVC(C=0.1, dual=False, class_weight='balanced')
svc_clf.fit(X, y)
y_pred = svc_clf.predict(X)
```

```python
# 計算錯誤分類的數量
err_num = (y_pred != y).sum()
print(f'錯誤分類的數量:{err_num}')
```

```python
# 計算模型準確度accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy:{accuracy:.4f}')
```

## F1-score
F1-score 是一種 分類模型評估指標，用來平衡 Precision（精確率） 和 Recall（召回率），特別適用於 類別不平衡 的情況。


*    Precision（精確率）
定義：當模型預測為 正類別 時，實際為正類別的比例。TP/(TP+FP誤報)
*   Recall（召回率）
定義：實際為 正類別 的樣本中，模型正確預測為正類別的比例。TP/(TP+FN漏報)



```python
#計算有加權的 F1-score (weighted:根據每個類別的樣本數，加權計算 F1-score)

from sklearn.metrics import f1_score
f1 = f1_score(y, y_pred, average='weighted')

print(f'F1_score:{f1:.4f}')
```

```python
# 預測未知寶可夢的 Type1
data1 = [[100,75]]
data1_pred = svc_clf.predict(data1)
label = le.inverse_transform(data1_pred)
print(f'預測的寶可夢種類為:{label}')
```
