```python
# TODO
# Importing the dataset
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/TQC+/data/wine.csv',header=None) #header=None 第一行第一列也要列進去
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values

print("X:",X)
print("y:",y)
print("X_shape:",X.shape)
print("y_shape:",y.shape)
```

```python

# TODO
# 75% 為訓練資料集、25%為測試資料集，random_state=5

from sklearn import model_selection
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 5)
#random_state=5,確保這個隨機分配的過程是可重現的

```

```python

from sklearn.tree import DecisionTreeClassifier

#TODO
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
# TODO
# compute accuracy of the classifier計算分類器的精確度
pred = classifier.predict(X_test)
accuracy = np.sum(pred == y_test) / len(y_test)
# pred == y_test會是一個布林值,正確預測為True=1，np.sum則是加總所有為true=1的總量
# len(y_test)則是看總數量
accuracy = round(accuracy,2)
# 用round計算至小數點第二位
print("Accuracy of the classifier =",accuracy)

```
